```python
#Steps to import data from a file on the computer:
#1. Import required library, in this case panda, provide path, syntax to read the file - df = pd.read(path, header = none optional)

import pandas as pd
path = "C:/Users/shirisha_valluri/Desktop/Desktop items/Coursera/imports-85.data"
df = pd.read_csv(path,header = None)
```

```python
#to get a sample of 10 rows from the bottom use df.tail, df.head for first few rows
df.tail(10)
```

```python
# create headers list
# if headers are missing, its possible to import headers from an online file or a list, similar to the one below.
# all the column names are separated by commas and are enclosed in double quotes. 
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
# printing headers 
print("headers\n",headers)
```

```python
# adding headers 
df.columns= headers
df.head(10)
```

```python
#dropping missing values of price and the corresponding row , if the subset is not mentioned other rows/columns would be affected
df.dropna(subset=["price"], axis=0)
```
